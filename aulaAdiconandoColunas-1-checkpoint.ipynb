```python
import pandas as pd
```

```python
nomes = ['Carla', 'João', 'Rafael', 'Ana', 'Lais']
cidade = ['São Paulo', 'Recife', 'Salvador', 'Porto Alegre', 'Rio de Janeiro']
matematica = [6, 7, 8, 9, 10]
biologia = [5, 3, 3, 4, 10]
data = {'Nome': nomes,
        'Cidade': cidade,
        'Matemática': matematica,
        'Biologia': biologia
    
}
df = pd.DataFrame(data)
```

## Função Apply

```python
def situacao(media):
    if media >= 6:
        return 'Aprovado'
    else:
        return 'Reprovado'
```

```python
df['Resultado_Final'] = df['Média'].apply(situacao)
```
